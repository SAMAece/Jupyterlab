```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Br fractions
x = np.linspace(0,1,13)

# Lattice parameters of mixing
a = pd.Series([24.9981813575002434, 24.8706342423675935, 24.7430871272349435, 24.6155400121022900, 24.4879928969696401,
              24.3604457818369902, 24.2328986667043367, 24.1053515515716867, 23.9778044364390368,
              23.8502573213063869, 23.7227102061737369, 23.5951630910410834, 23.4676159759084335])

b = pd.Series([25.4549722329332830, 25.3424933063235613, 25.2300143797138361, 25.1175354531041144, 25.0050565264943927,
              24.8925775998846710, 24.7800986732749493, 24.6676197466652241, 24.5551408200555024,
              24.4426618934457807, 24.3301829668360554, 24.2177040402263337, 24.105225113616612])

c = pd.Series([25.6514895484707246, 25.5405473895976556, 25.4296052307245795, 25.3186630718515069, 25.2077209129784343,
              25.0967787541053617, 24.9858365952322927, 24.8748944363592202, 24.7639522774861476,
              24.6530101186130750, 24.5420679597400024, 24.4311258008669299, 24.3201836419938573])

# merge three axis parameters
abc = pd.concat([a,b,c],axis=1)

# reduce parameter to one crystal
abc0 = abc/4

# rename columns
abc0.columns = ["a","b","c"]
```
