```python
#0和12还没算完

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

bg = pd.DataFrame({"Bandgap":[2.152, 2.175, 2.200, 2.277, 2.246, 2.286, 2.350, 2.383, 2.422, 2.472, 2.531],
"VBM":[1.065, 1.025, 0.965, 0.878, 0.849, 0.766, 0.664, 0.603, 0.511, 0.408, 0.317],
"CBM":[3.217, 3.200, 3.165, 3.155, 3.095, 3.052, 3.015, 2.986, 2.933, 2.880, 2.848,]})

bg["x"] = np.linspace(1/12,11/12,11)

plt.plot(bg.x,bg.Bandgap,marker='o',label = "Bandgap")
plt.plot(bg.x,bg.VBM,marker='o',label = "VBM")
plt.plot(bg.x,bg.CBM,marker='o',label = "CBM")
plt.xlabel("Br content x in $FAPb(Br_xI_{(1-x)})_3$")
plt.ylabel("Band energy (eV)")
plt.legend()
```

```python
# delete the error point Br4I8
bg = bg.drop(index=3)

EI = bg.Bandgap[0]
EBr = bg.Bandgap[10]

# calculating bandgap bowing parameter b
bg["b"] = EI / bg.x + EBr / (1-bg.x) - bg.Bandgap / (bg.x*(1-bg.x))

# compute mean value
b_mean = bg.b.mean()

print("b mean value is ", b_mean)
```
