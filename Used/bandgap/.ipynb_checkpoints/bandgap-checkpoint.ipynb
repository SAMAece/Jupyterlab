```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

bg = pd.DataFrame({"Bandgap":[1.195,1.547,1.561,1.575,1.643,1.599,1.624,1.672,1.698,1.722,1.760,1.806,1.806],
"VBM":[1.289,1.338,1.300,1.243,1.160,1.134,1.059,0.964,0.905,0.820,0.723,0.640,0.235],
"CBM":[2.484,2.885,2.861,2.818,2.803,2.733,2.682,2.636,2.603,2.542,2.483,2.445,2.041]})

x = np.linspace(0,1,13)


plt.plot(x,bg.Bandgap,marker='o',label = "Bandgap")
plt.plot(x,bg.VBM,marker='o',label = "VBM")
plt.plot(x,bg.CBM,marker='o',label = "CBM")
plt.xlabel("Br content")
plt.ylabel("Band energy (eV)")
plt.legend()
```
