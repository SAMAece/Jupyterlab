```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

bg = pd.DataFrame({"Bandgap":[1.523,1.547,1.561,1.575,1.643,1.599,1.624,1.672,1.698,1.722,1.760,1.806,1.844],
"VBM":[1.392,1.338,1.300,1.243,1.160,1.134,1.059,0.964,0.905,0.820,0.723,0.640,0.530],
"CBM":[2.915,2.885,2.861,2.818,2.803,2.733,2.682,2.636,2.603,2.542,2.483,2.445,2.374]})

bg["x"] = np.linspace(0,1,13)

plt.plot(bg.x,bg.Bandgap,marker='o',label = "Bandgap")
plt.plot(bg.x,bg.VBM,marker='o',label = "VBM")
plt.plot(bg.x,bg.CBM,marker='o',label = "CBM")
plt.xlabel("Br content x in FAPb(Br$_x$I$_{(1-x)}$)$_3$")
plt.ylabel("Band energy (eV)")
plt.legend()
```

```python
# delete the error point Br4I8
bg = bg.drop(index=4)

EI = bg.Bandgap[0]
EBr = bg.Bandgap[12]

# calculating bandgap bowing parameter b
bg["b"] = EI / bg.x + EBr / (1-bg.x) - bg.Bandgap / (bg.x*(1-bg.x))

# compute mean value
b_mean = bg.b.mean()

print("b mean value is ", b_mean)
```
