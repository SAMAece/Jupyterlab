```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# input Ex (from Br0 to Br12)
Ex = [-3666.65366132,
     -3.6754420815E+03,
     -3.6839027452E+03,
     -3.6931964747E+03,
     -3.7018743503E+03,
     -3.7108255544E+03,
     -3.7198376263E+03,
     -3.7291860202E+03,
     -3.7380176509E+03,
     -3.7476733273E+03,
     -3.7569261148E+03,
     -3.7659887746E+03,
     -3776.00723982
     ]

# create dataframe
df = pd.DataFrame({"x":np.linspace(0,1,13),"Ex": Ex})

# Ex per atom
df.Ex = df.Ex / 768

# set EBr and EI
EBr = df.Ex.iloc[-1]
EI = df.Ex.iloc[0]
```

```python
# show free energy per atom of mixing perovskite
plt.plot(df.x,df.Ex)
plt.xlabel('Br content')
plt.ylabel('Free energy (eV/atom)')
```

```python
# calculate dH
df["dH"] = df.Ex - df.x * EBr - (1 - df.x) * EI

# turn eV into meV
df.dH = 1000*df.dH
```

```python
# fitting dH points
coefficients = np.polyfit(df.x,df.dH,2)
a = coefficients[0]
b = coefficients[1]
c = coefficients[2]
def fit(x):
    return a*x**2+b*x+c

fit_y = fit(np.linspace(0,1,100))
```

```python
# draw formation enthalpy per atom
plt.scatter(df.x,df.dH, marker='o', color = "green")
plt.plot(np.linspace(0,1,100),fit_y)
plt.xlabel('Br content')
plt.ylabel('Formation Enthalpy (meV/atom)')

```
